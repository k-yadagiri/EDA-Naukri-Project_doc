```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options as EdgeOptions
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

# Setup WebDriver
service = EdgeService(EdgeChromiumDriverManager().install())
options = EdgeOptions()
driver = webdriver.Edge(service=service, options=options)

job_title=[]
company_name= []
experience=[]
salary=[]
location=[]
skills = []
rating = []
qualification =[]
reviews = []
for j in range(1, 36): 
    url = f"https://www.naukri.com/data-science-jobs-{j}?k=data+science"
    driver.get(url)
    time.sleep(3) 
    soup = BeautifulSoup(driver.page_source, "html.parser")
    for i in soup.find_all("div", class_="cust-job-tuple layout-wrapper lay-2 sjw__tuple"):      # continer 
        t = i.find("div",class_="row1")
        if t:
            job_title.append(t.text)
        else:
            job_title.append(np.nan)
        
        c = i.find("a",class_="comp-name mw-25")
        if c:
            company_name.append(c.text)
        else:
            company_name.append(np.nan)

        exp = i.find("span",class_="expwdth")
        if exp:
            experience.append(exp.text)
        else:
            experience.append(np.nan)

        sal = i.find("span",class_ = "Not disclosed") 
        if sal:
            salary.append(sal.text)
        else:
            salary.append(np.nan)

        loc = i.find("span",class_ = "locWdth")  
        if loc:
            location.append(loc.text)
        else:
            location.append(np.nan)

        sk = i.find("div",class_ = "row5") 
        if sk:
            skills.append(sk.text)
        else:
            skills.append(np.nan)

        rat = i.find("span",class_ = "main-2")
        if rat:
            rating.append(rat.text)
        else:
            rating.append(np.nan)

        qual = i.find("span",class_="job-desc ni-job-tuple-icon ni-job-tuple-icon-srp-description") 
        if qual:
            qualification.append(qual.text)
        else:
            qualification.append(np.nan)
        rev = i.find("a",class_="review ver-line") 
        if rev:
            reviews.append(rev.text)
        else:
            reviews.append(np.nan)

        
        
```

```python
dic={"JOB_TITLE":job_title,"COMPANY_NAME":company_name,"EXPERIENCE":experience,"SALARY":salary,
     "LOCATION":location,"SKILLS":skills,"RATING":rating,"QUALIFICATION":qualification,"reviews":reviews}
```

```python
df=pd.DataFrame(dic)
```
